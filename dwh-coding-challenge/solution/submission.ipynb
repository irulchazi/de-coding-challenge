```python
import glob
import json
import os

import pandas as pd
```

```python
path_json_accounts = '../data/accounts/' 
```

```python
json_accounts = glob.glob(os.path.join(path_json_accounts, '*.json'))
```

```python
accounts = pd.DataFrame()

for list in json_accounts:
    with open(list, 'rb') as file:
        data = json.load(file)
        if 'set' in data:
            data['set'] = str(data['set'])
            df_temp = pd.DataFrame.from_dict([data])
            accounts = accounts.append(df_temp)
        elif 'data' in data:
            data['data'] = str(data['data'])
            df_temp = pd.DataFrame.from_dict([data])
            accounts = accounts.append(df_temp)
```

```python
# to handle unix timestamp, if omitted, pandas will converts ts column defaulting to 1970 
accounts['ts'] = (accounts['ts']).astype(int) // 10**3
accounts['ts'] = pd.to_datetime(accounts['ts'], unit='s')
```

```python
# run this if you want to convert datetime to UTC+7/Jakarta time
accounts['ts_jakarta'] = accounts['ts'].dt.tz_localize('utc').dt.tz_convert('Asia/Jakarta')
```

```python
# reorder column for more readability in different timezone side-by-side
accounts = accounts[['id', 'op', 'ts', 'ts_jakarta', 'set', 'data']]
accounts_original = accounts[['id', 'op', 'ts', 'set', 'data']]
```
